```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate

url = 'https://raw.githubusercontent.com/fabiobanyu/File_Tugas/refs/heads/main/Analisis%20Alokasi%20Waktu%20perminggu%20mahasiswa%20penggemar%20Anime%20dengan%20Metode%20Snowball%20Sampling%20data%20real.csv'

data_set = pd.read_csv(url)
data_set = data_set.dropna()

print(tabulate(data_set.head(), headers='keys', tablefmt='pretty'))
print(tabulate(data_set.tail(), headers='keys', tablefmt='pretty'))
```

```python
# Pilih kolom yang relevan
jam_kolom = "Alokasi Waktu yang Anda habiskan untuk menonton anime perminggunya (Dalam Jam)"  # Nama kolom

# Hitung jumlah setiap kategori
jumlah_jam = data_set[jam_kolom].value_counts()

# Barplot
plt.figure(figsize=(10, 6))
sns.barplot(x=jumlah_jam.index, y=jumlah_jam.values, palette='viridis')
plt.title('Distribusi Waktu Untuk menonton Anime dalam jam/minggu', fontsize=14)
plt.xlabel('Alasan', fontsize=12)
plt.ylabel('Jumlah', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(
    jumlah_jam.values,
    labels=jumlah_jam.index,
    autopct='%1.1f%%',
    colors=sns.color_palette('coolwarm', len(jumlah_jam)),
    startangle=140,
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Distribusi Waktu Untuk menonton Anime dalam jam/minggu', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Pilih kolom yang relevan
Genre_kolom = "Genre Favorit"  # Nama kolom

# Hitung jumlah setiap kategori
jumlah_genre_favorit = data_set[Genre_kolom].value_counts()

# Barplot
plt.figure(figsize=(10, 6))
sns.barplot(x=jumlah_genre_favorit.index, y=jumlah_genre_favorit.values, palette='plasma')
plt.title('Genre Favorit', fontsize=14)
plt.xlabel('Alasan', fontsize=12)
plt.ylabel('Jumlah', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(
    jumlah_genre_favorit.values,
    labels=jumlah_genre_favorit.index,
    autopct='%1.1f%%',
    colors=sns.color_palette('rainbow', len(jumlah_genre_favorit)),
    startangle=140,
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Genre Favorit', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
data_set['Alokasi Waktu yang Anda habiskan untuk menonton anime perminggunya (Dalam Jam)'] = (
    data_set['Alokasi Waktu yang Anda habiskan untuk menonton anime perminggunya (Dalam Jam)']
    .astype(str)  # Konversi ke string
    .str.replace(' Jam', '', regex=False)  # Hapus teks " Jam"
    .str.strip()  # Hapus spasi tambahan
)

data_set['Alokasi Waktu yang Anda habiskan untuk menonton anime perminggunya (Dalam Jam)'] = pd.to_numeric(
    data_set['Alokasi Waktu yang Anda habiskan untuk menonton anime perminggunya (Dalam Jam)'],
    errors='coerce'
)

data_set = data_set[data_set['Alokasi Waktu yang Anda habiskan untuk menonton anime perminggunya (Dalam Jam)'] <= 100]
data_set = data_set.dropna(subset=['Alokasi Waktu yang Anda habiskan untuk menonton anime perminggunya (Dalam Jam)'])

data_set['Genre Favorit'] = data_set['Genre Favorit'].str.title().replace('', None)

data_set = data_set.dropna(subset=['Genre Favorit'])

data_set['Alasan Suka Anime'] = data_set['Alasan Suka Anime'].fillna("Tidak Ada Alasan")

data_set["Panjang Alasan"] = data_set["Alasan Suka Anime"].apply(len)

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(
    data=data_set,
    x="Genre Favorit",
    y="Alokasi Waktu yang Anda habiskan untuk menonton anime perminggunya (Dalam Jam)",
    palette="viridis",
    order=data_set["Genre Favorit"].value_counts().index,
)
plt.title("Distribusi Waktu Menonton Berdasarkan Genre Favorit", fontsize=14)
plt.xlabel("Genre Favorit", fontsize=12)
plt.ylabel("Alokasi Waktu (Jam)", fontsize=12)
plt.xticks(rotation=45)
plt.ylim(0, 60)

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=data_set,
    x="Panjang Alasan",
    y="Alokasi Waktu yang Anda habiskan untuk menonton anime perminggunya (Dalam Jam)",
    hue="Genre Favorit",
    palette="viridis",
    alpha=0.8,
)
plt.title("Korelasi Panjang Alasan Suka Anime dan Waktu Menonton", fontsize=14)
plt.xlabel("Panjang Alasan (Jumlah Karakter)", fontsize=12)
plt.ylabel("Alokasi Waktu (Jam)", fontsize=12)
plt.legend(title="Genre Favorit", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.xlim(0, 20)
plt.ylim(0, 20)  
plt.show()
```
