```python
#Nomor 1
import networkx as nx
import matplotlib.pyplot as plt

class Graph:
    def __init__(self):
        self.graph = {}

    def add_vertex(self, vertex):
        if vertex not in self.graph:
            self.graph[vertex] = []

    def add_edge(self, u, v):
        if u in self.graph and v in self.graph:
            self.graph[u].append(v)
            self.graph[v].append(u)

    def display(self):
        for vertex in self.graph:
            print(f'{vertex} : {self.graph[vertex]}')

    def visualize(self):
        G = nx.Graph()

        for vertex in self.graph:
            G.add_node(vertex)
        
        for u in self.graph:
            for v in self.graph[u]:
                if u < v:
                    G.add_edge(u, v)

        pos = nx.spring_layout(G)
        nx.draw(G, pos, with_labels=True, node_color='purple', node_size=2000, font_size=15, font_weight='bold')
        plt.title('Graph Visualization')
        plt.show()


#graph
g = Graph()

#vertex
g.add_vertex(1)
g.add_vertex(2)
g.add_vertex(3)
g.add_vertex(4)
g.add_vertex(5)
g.add_vertex(6)
g.add_vertex(7)
g.add_vertex(8)
g.add_vertex(9)
g.add_vertex(10)

#edge
g.add_edge(1, 10)
g.add_edge(2, 9)
g.add_edge(3, 8)
g.add_edge(4, 7)
g.add_edge(5, 6)
g.add_edge(1, 5)
g.add_edge(2, 6)
g.add_edge(3, 7)
g.add_edge(4, 8)
g.add_edge(5, 10)
g.add_edge(6, 9)
g.add_edge(7, 8)
g.add_edge(8, 10)
g.add_edge(9, 1)
g.add_edge(10, 2)
g.add_edge(1, 3)
g.add_edge(2, 4)
g.add_edge(3, 5)
g.add_edge(4, 6)
g.add_edge(5, 7)
g.add_edge(6, 8)
g.add_edge(7, 9)
g.add_edge(8, 1)
g.add_edge(9, 2)
g.add_edge(10, 3)

print("Graph adjacency list:")
g.display()

g.visualize()
```

```python
#Nomor 2
import networkx as nx
import matplotlib.pyplot as plt

class GraphMatrix:
    def __init__(self, num_vertices):
        self.num_vertices = num_vertices
        self.matrix = [[0] * num_vertices for _ in range(num_vertices)]
        self.node_map = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4, 'F': 5}
        self.reverse_node_map = {v: k for k, v in self.node_map.items()}

    def add_edge(self, u, v):
        u_index = self.node_map[u]
        v_index = self.node_map[v]
        self.matrix[u_index][v_index] = 1
        self.matrix[v_index][u_index] = 1 

    def display(self):
        print("Adjacency Matrix:")
        for row in self.matrix:
            print(row)

    def visualize(self):
        G = nx.Graph()

        for label in self.node_map:
            G.add_node(label)

        for i in range(self.num_vertices):
            for j in range(i, self.num_vertices):
                if self.matrix[i][j] == 1:
                    u = self.reverse_node_map[i]
                    v = self.reverse_node_map[j]
                    G.add_edge(u, v)

        pos = nx.spring_layout(G)  # Layout untuk penempatan nodes
        nx.draw(G, pos, with_labels=True, node_color='blue', node_size=2000, font_size=15, font_weight='bold')
        plt.title('Graph Visualization')
        plt.show()

g = GraphMatrix(6)

g.add_edge('A', 'B')
g.add_edge('A', 'C')
g.add_edge('A', 'D')
g.add_edge('B', 'A')
g.add_edge('B', 'E')
g.add_edge('C', 'A')
g.add_edge('C', 'F')
g.add_edge('D', 'A')
g.add_edge('E', 'B')
g.add_edge('F', 'C')

g.display()

g.visualize()
```

```python
#Nomor 3
import networkx as nx
import matplotlib.pyplot as plt

class Graph:
    def __init__(self):
        self.graph = {} 

    def add_vertex(self, vertex):
        if vertex not in self.graph:
            self.graph[vertex] = []

    def add_edge(self, u, v):
        if u in self.graph and v in self.graph:
            self.graph[u].append(v)
            self.graph[v].append(u)

    def display(self):
        for vertex in self.graph:
            print(f'{vertex} : {self.graph[vertex]}')

g = Graph()

for i in range(8):
    g.add_vertex(i)

g.add_edge(0, 1)
g.add_edge(0, 3)
g.add_edge(0, 5)
g.add_edge(1, 2)
g.add_edge(1, 6)
g.add_edge(2, 3)
g.add_edge(2, 7)
g.add_edge(3, 4)
g.add_edge(4, 5)
g.add_edge(4, 7)
g.add_edge(5, 6)
g.add_edge(6, 7)

print("\nAdjacency List:\n")
g.display()

class GraphMatrix:
    def __init__(self, num_vertices):
        self.num_vertices = num_vertices
        self.matrix = [[0] * num_vertices for _ in range(num_vertices)]

    def add_edge(self, u, v):
        if u < self.num_vertices and v < self.num_vertices:
            self.matrix[u][v] = 1
            self.matrix[v][u] = 1

    def display(self):
        for row in self.matrix:
            print(row)

    def visualize(self):
        G = nx.Graph()
        for i in range(self.num_vertices):
            for j in range(self.num_vertices):
                if self.matrix[i][j] == 1:
                    G.add_edge(i, j)

        pos = nx.spring_layout(G)
        nx.draw(G, pos, with_labels=True, node_color='red', node_size=500, font_size=10, font_color='black')
        plt.title("Graph Visualization (Adjacency Matrix)")
        plt.show()

g_matrix = GraphMatrix(8)

g_matrix.add_edge(0, 1)
g_matrix.add_edge(0, 3)
g_matrix.add_edge(0, 5)
g_matrix.add_edge(1, 2)
g_matrix.add_edge(1, 6)
g_matrix.add_edge(2, 3)
g_matrix.add_edge(2, 7)
g_matrix.add_edge(3, 4)
g_matrix.add_edge(4, 5)
g_matrix.add_edge(4, 7)
g_matrix.add_edge(5, 6)
g_matrix.add_edge(6, 7)

print("\nAdjacency Matrix:\n")
g_matrix.display()

g_matrix.visualize()
```
