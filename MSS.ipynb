```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as  sns
from scipy.stats import skew

data_set = pd.read_excel("D:/FILE AND TASK/data_mss8.xlsx")
print(data_set.head())

mean_rating = data_set['rating'].mean()
sd_rating = data_set['rating'].std()

print("Rata-rata rating: ", mean_rating)
print("Standara Deviasi rating: ", sd_rating)
```

```python
uji = data_set['rating']
uji.describe()
```

```python
#histogram
plt.figure(figsize=(10, 5))
plt.hist(uji, bins=5, color='skyblue', edgecolor='black')
plt.axvline(mean_rating, color='red', linestyle='dashed', linewidth=1)
plt.title('Histogram Rating')
plt.xlabel('Rating')
plt.ylabel('Customer')
plt.legend(['Rata-rata', 'Rating'])
plt.show()
```

```python
# Create boxplot
plt.figure(figsize=(10, 8))
plt.boxplot(uji, vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Boxplot of Ratings')
plt.ylabel('Rating')
plt.show()
```
