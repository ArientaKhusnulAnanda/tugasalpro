```python
# 1. Buatlah sebuah Pandas DataFrame (sudah dibuat di langkah sebelumnya)

# 2. Informasi umum tentang DataFrame
info_umum = df.info()

# 3. Hitung jumlah karyawan dari masing-masing kota
jumlah_karyawan_per_kota = df['Kota'].value_counts()

# 4. Data orang di bidang Teknik dengan pengalaman kerja lebih dari 5 tahun
bidang_teknik = df[df['Pekerjaan'].str.contains("Engineer|Developer|IT Support|System Analyst|Administrator", case=False) &
                   (df['Pengalaman Kerja (tahun)'] > 5)]

# 5. Data orang dengan gaji lebih dari 12 juta dan memiliki sertifikasi Scrum atau Agile
# Asumsi sertifikasi Scrum atau Agile: baris di kolom "Sertifikasi" yang mengandung kata "Scrum" atau "Agile"
sertifikasi_scrum_agile = df[df['Sertifikasi'].str.contains("Scrum|Agile", case=False)]

# 6. Data Nama, Pekerjaan, dan Kota dari orang yang bergabung antara tahun 2015-2020
df['Tanggal Bergabung'] = pd.to_datetime(df['Tanggal Bergabung'])
periode_bergabung = df[(df['Tanggal Bergabung'].dt.year >= 2015) & (df['Tanggal Bergabung'].dt.year <= 2020)][
    ['Nama', 'Pekerjaan', 'Kota']]

# 7. Tambah kolom "Status"
df['Status'] = df['Umur'].apply(lambda x: "Lajang" if x <= 25 else "Menikah")

# 8. Simpan DataFrame yang dimodifikasi ke file CSV
output_file = "/mnt/data/dataset_karyawan.csv"
df.to_csv(output_file, index=False)

# Display results for verification
info_umum, jumlah_karyawan_per_kota, bidang_teknik, sertifikasi_scrum_agile, periode_bergabung, output_file
```
