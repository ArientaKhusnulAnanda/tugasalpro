### **Series**

Series dalam Pandas seperti sebuah kolom dalam tabel / Array satu dimensi yang menyimpan data jenis apa pun.

```python
a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar)
```

```python
print(myvar[0])
```

```python
#Menggunakan Index
a = [1, 7, 2]
myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)
```

```python
print(myvar["y"])
```

```python
#Menggunakan Dictionary
calories = {
    "day1": 420, 
    "day2": 380, 
    "day3": 390
    }
myvar = pd.Series(calories)
print(myvar)

#dictionary menjadi label.
```

- Untuk memilih hanya beberapa item dalam dictionary, gunakan index argumen dan tentukan hanya item yang ingin disertakan dalam Series. Contoh:

```python
calories = {
    "day1": 420, 
    "day2": 380, 
    "day3": 390
    }
myvar = pd.Series(calories, index = ["day1", "day2"])
print(myvar)
```

### **Data Frame**

Kumpulan data dalam Pandas biasanya berupa tabel multidimensi yang disebut DataFrames. Jika series seperti kolom, DataFrame adalah keseluruhan tabel.

```python
data = {
 "calories": [420, 380, 390],
 "duration": [50, 40, 45]
}
df = pd.DataFrame(data)
print(df)
```

```python
print(df.loc[0],"\n")
print(df.loc[[0, 1]])
```

### **Pandas Read CSV**

Cara sederhana untuk menyimpan kumpulan data besar adalah dengan menggunakan file CSV. File CSV berisi teks biasa dan merupakan format yang dapat dibaca oleh semua orang termasuk Pandas.Dalam contoh akan digunakan file CSV yang disebut 'data.csv'. data.csv. 

```python
url = "https://raw.githubusercontent.com/fabiobanyu/File_Tugas/refs/heads/main/Analisis%20Kebiasaan%20Mahasiswa%20tidak%20Memakai%20Helm%20Saat%20Berkendara%20(Responses)%20-%20Form%20Responses.csv"

data_frame = pd.read_csv(url)
print(data_frame.to_string())
```

### **Latihan Praktikum**

```python
import pandas as pd

mydataset = {
    'aset': ["rumah", "mobil", "motor"],
    'harga': [500000000, 200000000, 20000000]
}

myvar = pd.DataFrame(mydataset)
print(myvar)
```

```python
mydataset = {
    'aset': ["rumah", "mobil", "motor"],
    'harga': [500000000, 200000000, 20000000]
}

myvar = pd.DataFrame(mydataset, index=['a', 'b', 'c'])
print(myvar)
```

```python
mydataset = [
    {'rumah': 1, 'mobil': 2, 'motor': 3},
    {'rumah': 1, 'mobil': 1, 'motor': 1}
]

myvar = pd.DataFrame(mydataset)
print(myvar)
```

```python
mydataset = {
    'aset1': pd.Series([1, 2, 3]),
    'aset2': pd.Series([1, 1, 1])
}

myvar = pd.DataFrame(mydataset)
print(myvar)
```
